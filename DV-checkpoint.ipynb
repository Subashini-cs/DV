```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the Spotify dataset
spotify_data = pd.read_csv("C:/Users/subas/Downloads/spotify_dataset.csv")
spotify_data

```

```python
spotify_data.head()
```

```python
spotify_data.info()
```

```python
%matplotlib inline

f,ax = plt.subplots(figsize=(14,10))
sns.heatmap(spotify_data.corr(),annot = True,fmt = ".1f",ax = ax)
plt.show()
```

```python
# Step 3: Exploratory Data Analysis (EDA)

# Univariate Analysis
plt.figure(figsize=(12, 6))
sns.histplot(data=spotify_data, x='Popularity', kde=True, bins=20, color='skyblue')
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()

```

```python
# Bivariate Analysis
plt.figure(figsize=(10, 8))
sns.scatterplot(data=spotify_data, x='Energy', y='Danceability')
plt.title('Energy vs Danceability')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.show()


```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Number of Times Charted", y="Popularity", data=spotify_data, palette="viridis")
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.subplots()
spotify_data.Genre.value_counts()[:30].plot(ax=ax, kind = "pie")
ax.set_ylabel("")
ax.set_title("Top 30 most popular genres")
plt.show()
```

```python
index=range(1, 11)
plt.figure(figsize=(10, 6))
spotify_data.boxplot()
plt.title('Box Plot')
```

```python
plt.figure(figsize=(10, 10))
sns.violinplot(data=spotify_data, palette="Pastel1")

plt.title('Violin Plot')

# Display the plot
plt.show()
```

```python
sns.jointplot(x='Danceability', y='Energy', data=spotify_data, kind='scatter')
plt.suptitle('Joint Plot', y=1.02)
plt.show()
```
